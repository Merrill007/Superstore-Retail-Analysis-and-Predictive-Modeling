# Dataset

------

The dataset consist of 21 columns and a total of 9994 rows.

Here's an overview of the columns:

1. **Row ID**: A unique identifier for each row.
2. **Order ID**: A unique identifier for each order.
3. **Order Date**: Date when the order was placed.
4. **Ship Date**: Date when the order was shipped.
5. **Ship Mode**: The mode of shipment (e.g., Standard Class, Second Class).
6. **Customer ID**: A unique identifier for each customer.
7. **Customer Name**: Name of the customer.
8. **Segment**: Market segment (e.g., Consumer, Corporate).
9. **Country**: Country where the order was placed.
10. **City**: City where the order was placed.
11. **State**: State where the order was placed.
12. **Postal Code**: Postal code of the location.
13. **Region**: Geographic region (e.g., East, West, Central, South).
14. **Product ID**: A unique identifier for each product.
15. **Category**: Category of the product (e.g., Furniture, Office Supplies, Technology).
16. **Sub-Category**: More specific classification within a category (e.g., Chairs, Binders, Phones).
17. **Product Name**: Name of the product.
18. **Sales**: The amount of sales for the product.
19. **Quantity**: The quantity of the product ordered.
20. **Discount**: The discount applied to the product.
21. **Profit**: The profit gained from the sale.

# Predictive Modeling for Sales and Profits

------

```python
# Feature Selection
features_sales = ['Discount', 'Quantity', 'Ship Mode']  
features_profits = ['Discount', 'Quantity', 'Category']  

# Data Splitting
from sklearn.model_selection import train_test_split

X_sales = df[features_sales]
y_sales = df['Sales']

X_profits = df[features_profits]
y_profits = df['Profit']

# Split for Sales
X_train_sales, X_test_sales, y_train_sales, y_test_sales = train_test_split(X_sales, y_sales, test_size=0.2, random_state=42)

# Split for Profits
X_train_profits, X_test_profits, y_train_profits, y_test_profits = train_test_split(X_profits, y_profits, test_size=0.2, random_state=42)
```

```python
# For categorical variables like 'Ship Mode', you can use one-hot encoding. This converts each category into a binary column (0 or 1) indicating its presence or absence.
df_encoded = pd.get_dummies(df, columns=['Ship Mode'], drop_first=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Select numerical features for the model
numerical_features = ['Discount', 'Quantity'] 

# Define features and target variable
X = df[numerical_features]
y_sales = df['Sales']
y_profit = df['Profit']

# Split the data into training and testing sets
X_train_sales, X_test_sales, y_train_sales, y_test_sales = train_test_split(X, y_sales, test_size=0.2, random_state=42)
X_train_profit, X_test_profit, y_train_profit, y_test_profit = train_test_split(X, y_profit, test_size=0.2, random_state=42)

# Initialize and train the regression model for Sales
reg_sales = LinearRegression()
reg_sales.fit(X_train_sales, y_train_sales)

# Initialize and train the regression model for Profit
reg_profit = LinearRegression()
reg_profit.fit(X_train_profit, y_train_profit)

# Predict Sales and Profit
y_pred_sales = reg_sales.predict(X_test_sales)
y_pred_profit = reg_profit.predict(X_test_profit)

# Calculate Mean Squared Error for Sales and Profit
mse_sales = mean_squared_error(y_test_sales, y_pred_sales)
mse_profit = mean_squared_error(y_test_profit, y_pred_profit)

print(f"Mean Squared Error (Sales): {mse_sales}")
print(f"Mean Squared Error (Profit): {mse_profit}")

# Retrieve Coefficients and Intercept for Sales
coeff_sales = reg_sales.coef_
intercept_sales = reg_sales.intercept_

print(f"Coefficients (Sales): {coeff_sales}")
print(f"Intercept (Sales): {intercept_sales}")

# Retrieve Coefficients and Intercept for Profit
coeff_profit = reg_profit.coef_
intercept_profit = reg_profit.intercept_

print(f"Coefficients (Profit): {coeff_profit}")
print(f"Intercept (Profit): {intercept_profit}")
```

```python
# Generate predictions for Sales and Profit
sales_predictions = reg_sales.predict(X) 
profit_predictions = reg_profit.predict(X) 

# Create scatter plot for Sales
plt.figure(figsize=(10, 5))
plt.scatter(y_sales, sales_predictions, color='blue', alpha=0.5)
plt.title('Actual vs. Predicted Sales')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.show()

# Create scatter plot for Profit
plt.figure(figsize=(10, 5))
plt.scatter(y_profit, profit_predictions, color='red', alpha=0.5)
plt.title('Actual vs. Predicted Profit')
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.show()
```

Regarding the sales model:

- The model's accuracy in forecasting sales is indicated by the Mean Squared Error (MSE). Better performance is indicated by lower values. The MSE of 284008.57 provides a rough indication of the prediction accuracy but also indicates that there is potential for improvement.


- Coefficients: For sales, the weights given to the "Discount" and "Quantity" attributes are represented by coefficients. Discount has a bigger impact on sales in this situation than quantity does. For example, keeping the quantity constant, a one-unit increase in "Discount" equates to a 29.21-unit rise in "Sales." Similar to this, keeping the "Discount" constant, a one-unit increase in "Quantity" equates to a 55.12-unit rise in "Sales."
